$f(x) = -12 x^4 sin(cos(x)) - 18x^3+5x^2 + 10x - 30$

_Импортируем пакеты, необходимые для решения._

```python
import sympy as sp
from sympy.abc import x, y
from sympy import solve, diff
from sympy import symbols
from sympy.plotting import plot
```

_Для начала_
### ***4. Построим график:***

```python
y = sp.sin(x) ** 2 - sp.cos(x) ** 2
gr = plot(y, show=False)
gr.show()
```

 #### *График функции имеет период*
 ## $$\pi$$

*Найдём корни уравнения методом*   **solve() :**

```python
x = sp.Symbol('x')
y = sp.sin(x) ** 2 - sp.cos(x) ** 2
solve(y , x, dict=True)
```

 ### ***1. Корни уравнения равны :***
 ### $x = \pi/4 \pm n \pi/2$
\
\
*Найдем интервалы, на которых фукция возрастает :*

*Определим первую производную методом*  **diff()**. 

```python
y0 = diff(y)
y0
```

 *Первая производная равна :*
### $y'(x) = 4sin(x) cos(x)$
 *Функция возрастает на интервалах, где производная > 0, и убывает на интевалах, где производная < 0.*

```python
solve(y0 > 0 , x, dict=True)
```

 ### ***2. Функция возрастает на интервалах :***
 ### $ 0 \pm n \pi<x< \pi/2 \pm n \pi$
\
\
*Найдем интервалы, на которых фукция убывает :*

```python
solve(y0 < 0 , x, dict=True)
```

 ### ***3. Функция убывает на интервалах :***
 ### $ \pi/2 \pm n \pi<x< \pi \pm n \pi$
\
\
*Вычислим экстремумы : Здесь производная равна нулю.*

```python
solve(y0, x, dict=True)
```

 ### ***5. Экстремумы уравнения равны :***
 ### $x =  0 \pm n \pi/2$
\
\
*Найдем интервалы, на которых функция > 0 :*

```python
solve(y > 0 , x, dict=True)
```

 ### ***6. Функция > 0 на интервалах :***
 ### $ \pi/4 \pm n \pi<x< 3\pi/4 \pm n \pi$
\
\
*Найдем интервалы, на которых функция < 0 :*

```python
solve(y < 0 , x, dict=True)
```
